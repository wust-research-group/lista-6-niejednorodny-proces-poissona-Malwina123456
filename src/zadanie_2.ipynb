```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def odwrotna(x):
    return 1 / (x + 1)
```

```python
def inv_poiss(T):
    S = [0]
    t = 0
    I = 0
    while True:
        u = np.random.uniform(0, 1)
        n = odwrotna(u)
        t += n
        if t <= T:
            I += 1
            S.append(t)
        else:
            break
    return S
```

```python
T = 20
t = inv_poiss(T)

plt.step(t, range(len(t)))
plt.show()
```
